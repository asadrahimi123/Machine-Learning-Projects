```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
a = [0,2,4,7,9,10]
b = [0,1,4,5,8,9]

plt.figure(figsize=(5,5),dpi=100)
plt.plot(a,b,'k^--',label='2x')

ar = np.arange(0,4.5,0.5)
plt.plot(ar[:6],ar[:6]**2,'p-',label='labeles', color='gray')
plt.plot(ar[4:],ar[4:]**2,'r--^')
plt.title('it is good',fontsize=20)
plt.xlabel('scor',fontdict={'fontname':'Comic Sans MS','color':'blue','fontsize':15})

plt.savefig('myfig.png',dpi=250)
```

```python
data = {
    'criket_Bat':['sg','bdm','ss','gm'],
    'mrp':[2000,2200,2400,2700],
    'weight_Grams':[1100,1000,1350,1390]
}
dataFrame = pd.DataFrame(data)
plt.plot(dataFrame['mrp'],dataFrame['weight_Grams'])
plt.show()
```

```python
data = {
    'name':['Ali','Ahmad','Reza'],
    'age':[20,22,19],
    'salary':[20000,28000,13000]
}
dataFrame = pd.DataFrame(data)
plt.plot(dataFrame['age'],dataFrame['salary'])
plt.grid()
plt.xlabel('Age of each employee')
plt.ylabel('Salary of each employee')
plt.title('Age and Salary of employees')
plt.show()
```

```python
a = np.random.randint(1,10,4)
b = [2,6,8,9]
c = [1,3,5,7]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')

ax.legend(loc='upper center')

plt.title('Frequency of a Signal')

plt.show()
```

```python
a = np.random.randint(1,10,4)
b = [2,6,8,9]
c = [1,3,5,7]

fig = plt.figure()
ax = plt.subplot()

ax.plot(a,b,'k--',label='Frequency')
ax.plot(a,c,'k:',label='Periods')

legend = ax.legend(loc='upper center',fontsize='x-small')
legend.get_frame().set_facecolor('blue')

plt.title('Frequency of a Signal')

plt.show()
```

```python
names = ['Ali','Ahmad','Mahdi','Jawad']
score = [97,88,91,77]

result = plt.bar(names,score)

patterns = ['/','*','o','+']

for res in result:
    res.set_hatch(patterns.pop())

plt.title('Score of Student')
plt.xlabel('Student')
plt.ylabel('Score')

plt.show()
```

```python
runsers = ['Ali','Mahdi','Abdullah']
runs = [90,94,99]

plt.pie(runs,labels=runsers,autopct='%1.2f%%')

plt.title('Run scored with each player')

plt.show()
```

```python
months = ['may','jon','des']
rain = [90,94,99]

plt.plot(months,rain)

plt.xlabel('months')
plt.ylabel('rain (mm)')

plt.title('rainning per month')

plt.show()
```

```python
team1_score = [25,34,55,51,41,31]
team2_score = [55,21,50,41,21,18]
scoreRange = [5,10,15,20,25,30]

plt.scatter(team1_score, scoreRange, color='r')
plt.scatter(team2_score, scoreRange, color='b')

plt.title('Score of two teams')
plt.xlabel('Team Score')
plt.ylabel('Score Range')

plt.show()
```
